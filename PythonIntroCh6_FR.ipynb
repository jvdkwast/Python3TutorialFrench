## 6.2 La boucle `for`
En gros, la boucle `for` fait quelque chose pour chaque valeur d'une liste. La façon dont elle est présentée est un peu confuse, mais en fait elle est très basique. Voici un exemple de code:

````python
# Exemple de boucle 'for'
# D'abord, créez une liste à parcourir en boucle:
NouvelleListe = [45, 'mange moi', 90210, "Le jour est venu, dit le morse, \
de parler de beaucoup de choses", -67]

# créer la boucle:
# parcourir newList, et mettre séquentiellement chaque bit d'information
# dans la variable valeur, et exécute la boucle
for valeur in NouvelleListe:
    print(valeur)
````

Comme vous le voyez, lorsque la boucle s'exécute, elle parcourt toutes les valeurs de la liste mentionnée après `in`. Elle les place ensuite dans `valeur`, et exécute la boucle, pour chaque différent élément de valeur. Voyons encore une fois, dans un appel classique de cheerleading que nous connaissons tous:

````python
# Program de cheerleading
mots = input("Qui choisissez-vous? ")

for lettre in mots:
    appel = "Dites " + lettre + "!"
    print(appel)
    print(lettre + "!")

print("Comment ça se dit?")
print(mots + "!")
````

Voici ce que vous venez retenir:
* Comme vous le voyez, les chaînes de caractères (rappelez-vous - les chaînes de caractères sont des lignes de texte) sont juste des listes avec beaucoup de caractères.
* Le programme a parcouru chacune des lettres (ou valeurs) de mots, et il les a affiché à l'écran.<br>

Et c'est tout ce qu'il y a à faire avec la boucle.

## 6.4 Notre premier 'Jeu'
Quel sera notre premier exemple de programme? Que diriez-vous d'un jeu d'aventure textuel (très) simple ? Ça a l'air amusant ! Il ne concernera qu'une seule pièce d'une maison, et sera extrêmement simple. Il y aura cinq objets et une porte. Dans l'un des cinq objets se trouve la clé de la porte. Vous devez trouver la clé, puis ouvrir la porte. Je vais d'abord donner une version en français, puis en Python:<br>
```
#Version francçaise de notre "jeu"

Indiquer à l'ordinateur la fonction de notre menu

Afficher un message de bienvenue, montrant une description de la pièce.
Nous allons donner au joueur six choses à regarder : une plante en pot, un tableau,\
un vase, un abat-jour, une chaussure et la porte

Indiquer à l'ordinateur que la porte est verrouillée
Indiquer à l'ordinateur où se trouve la clé

Présentez un menu qui vous indique les objets que vous pouvez "manipuler":
    Il vous donnera les six options
    Il posera la question "que voulez-vous regarder?"

si l'utilisateur veut regarder:
    la plante en pot:
        Si la clé y est, donner la clé au joueur
        sinon, lui dire qu'elle n'est pas là
    tableau:
        même chose que ci-dessus
    etc.
    porte:
        Si le joueur a la clé, le laisser ouvrir la porte
        Sinon, lui dire de chercher d'avantage

Donnez au joueur un message de félicitations, pour avoir terminé le jeu.
```
<br>
A partir ceci, nous pouvons écrire un vrai programme. Vous êtes prêts ? C'est parti:

````python
# JEU D'AVENTURE TEXTUEL

#la fonction menu:
def menu(list, question):
    for entry in list:
        print(1 + list.index(entry),end="")
        print (") " + entry)

    return eval(input(question)) - 1

#Donnez à l'ordinateur quelques informations de base sur la pièce:
objets = ["plante en pot","tableau","vase","abat-jour","chaussure","porte"]

#La clé se trouve dans le vase (ou l'entrée numéro 2 dans la liste ci-dessus):
localisationclé = 2

#Vous n'avez pas trouvé la clé:
clétrouvée = 0

boucle = 1

#Donnez le texte d'introduction:
print("La nuit dernière, vous vous êtes endormi dans le confort de votre propre maison.")

print("Maintenant, vous vous retrouvez enfermés dans une pièce. Vous ne savez pas comment")
print("vous êtes arrivé là, ou quelle heure il est. Dans la pièce, vous pouvez voir")
print(len(objets), "objets:")
for x in objets:
    print(x)
print("")
print("La porte est fermée, y aurait-il une clé quelque part?")
#Faites fonctionner votre menu et le programme jusqu'à ce que vous trouviez la clé:
while boucle == 1:
    choix = menu(objets,"Que voulez-vous inspecter? ")
    if choix == 0:
        if choix == localisationclé:
            print("Vous avez trouvé une petite clé dans la plante en pot.")

            print("")
            clétrouvée = 1
        else:
            print("Vous n'avez rien trouvé dans la plante en pot.")
            print("")
    elif choix == 1:
        if choix == localisationclé:
            print("Vous avez trouvé une petite clé derrière le tableau.")
            print("")

            clétrouvée = 1
        else:
            print("Vous n'avez rien trouvé derrière le tableau.")
            print("")
    elif choix == 2:
        if choix == localisationclé:
            print("Vous avez trouvé une petite clé dans le vase.")
            print("")
            clétrouvée = 1
        else:
            print("Vous n'avez rien trouvé dans le vase.")

            print("")
    elif choix == 3:
        if choix == localisationclé:
            print("Vous avez trouvé une petite clé dans l'abat-jour.")
            print("")
            clétrouvée = 1
        else:
            print("Vous n'avez rien trouvé dans l'abat-jour.")
            print("")

    elif choix == 4:
        if choix == localisationclé:
            print("Vous avez trouvé une petite clé dans la chaussure.")
            print("")
            clétrouvée = 1
        else:
            print("Vous n'avez rien trouvé dans la chaussure.")
            print("")
    elif choix == 5:
        if clétrouvée == 1:
            boucle = 0
            print("Vous mettez la clé, vous la tournez, et vous entendez un clic")

            print("")
        else:
            print("La porte est fermée, vous devez trouver une clé.")
            print("")

# Rappelez-vous qu'une barre oblique inverse poursuit
# le code sur la ligne suivante
print("La lumière pénètre dans la pièce lorsque \
vous ouvrez la porte de votre liberté.")
````

Eh bien, un jeu très simple, mais amusant. Ne vous laissez pas décourager par la quantité de code, 53 des lignes ne sont que des instructions "si", ce qui est la chose la plus facile à lire (une fois que vous avez compris l'indentation).
